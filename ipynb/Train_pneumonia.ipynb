2. Train the model

```python
import os
from ultralytics import YOLO

# Path to DATASET
DATA_DIR = '\Documents\Intro to AI\Pneumonia_Dectection\data\pnemonia_data'

# Load a model
model = YOLO("yolov8n-cls.pt")  # load a pretained model

# Use the model
results = model.train(data=DATA_DIR, epochs=10, imgsz=224, batch= 32)  # train the model
```

```python
import os
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

# Check the current working directory
print("Current Working Directory:", os.getcwd())

# Define the file path
results_path = './runs/classify/train6/results.csv'

# Convert the relative path to an absolute path
absolute_results_path = os.path.abspath(results_path)
print("Absolute Path:", absolute_results_path)

# Check if the file exists
if os.path.exists(absolute_results_path):
    results = pd.read_csv(absolute_results_path)
    
    # Plotting code
    plt.figure()
    plt.plot(results['                  epoch'], results['             train/loss'], label='train loss')
    plt.plot(results['                  epoch'], results['               val/loss'], label='val loss', c='red')
    plt.grid()
    plt.title('Loss vs epochs')
    plt.ylabel('loss')
    plt.xlabel('epochs')
    plt.legend()

    plt.figure()
    plt.plot(results['                  epoch'], results['  metrics/accuracy_top1'] * 100)
    plt.grid()
    plt.title('Validation accuracy vs epochs')
    plt.ylabel('accuracy (%)')
    plt.xlabel('epochs')
    plt.show()
    # No need for plt.show() in Jupyter Notebook

else:
    print("File not found:", absolute_results_path)
```

```python
from ultralytics import YOLO
import os
from pathlib import Path
import numpy as np


model = YOLO('/Documents/Intro to AI/Pneumonia_Dectection/runs/classify/train6/weights/best.pt')  # load a custom model

directory = r"D:\Documents\Intro to AI\Pneumonia_Dectection\DATASET\chest_xray\test\PNEUMONIA"
allprob = 0
count = 0
files = Path(directory).glob("*")
for filename in files:
    results = model(filename)  # predict on an image
    names_dict = results[0].names
    probs = results[0].probs.data.tolist()
    print(names_dict)
    print(probs)
    count = count + 1
    allprob = allprob + probs[2]

print(allprob/count)
```
